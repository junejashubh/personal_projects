```python
import pandas as pd
import numpy as np

```

```python
# Read the CSV file
df = pd.read_csv('/Users/shubhamjuneja/vscode/personal_projects/neet_score_card/op/all_neet_marks_2607.csv')

# Group the data by the 'state' column
grouped = df.groupby('state')

# Convert the grouped data into a list of DataFrames
grouped_list = [group for _, group in grouped]

# Calculate the number of groups per part
num_parts = 4
groups_per_part = len(grouped_list) // num_parts

# Split the data into parts
parts = [pd.concat(grouped_list[i*groups_per_part:(i+1)*groups_per_part]) for i in range(num_parts)]

# Handle the remainder
if len(grouped_list) % num_parts != 0:
    remainder = pd.concat(grouped_list[num_parts*groups_per_part:])
    parts[-1] = pd.concat([parts[-1], remainder])

# Save each part as a separate CSV file
```

```python
for i in parts:
    print(len(i))
```

```python
for i, part in enumerate(parts):
    part.to_csv(f'/Users/shubhamjuneja/vscode/personal_projects/neet_score_card/op/part_{i+1}.csv', index=False)

print("Data split into 4 parts successfully.")
```

```python
state_wise_avg_marks = df.groupby(['state']).agg({'Marks':'mean'}).reset_index()
# state_wise_avg_marks['India_avg'] = np.mean(df.Marks)
# state_wise_avg_marks['India_median'] = np.median(df.Marks)
#state_wise_avg_marks.to_csv('state_avg.csv',index=False)
state_wise_avg_marks.rename(columns = {'Marks':'State_avg'},inplace = True)

```

```python
city_avg = df.groupby(['state','city']).agg({'Marks':'mean','center_num': pd.Series.nunique,'center_name':'count'}).reset_index()
city_avg = city_avg.merge(state_wise_avg_marks,on='state',how='left')
city_avg.rename(columns = {'Marks':'city_avg',
                           'center_num':'num_center',
                           'center_name':'num_students'},inplace = True)
city_avg.to_csv('city_avg.csv',index=False)
city_avg
```

```python
bins = [-200,0,100, 200, 300, 400, 500,550, 600, 650, 700, float('inf')]
labels = ['-200-0','0-100','100-200', '200-300', '300-400', '400-500', '500-550','550-600', '600-650', '650-700', '>700']

# Create a new column 'Marks_bin' with the bin labels
df['Marks_bin'] = pd.cut(df['Marks'], bins=bins, labels=labels, right=False)

# Group by 'state', 'city', and 'Marks_bin', and count the number of entries in each bin

```
